5. Kiểm tra năm nhuận
Nhập vào một năm bất kỳ. Kiểm tra xem năm đó có phải năm nhuận hay không theo quy tắc:

Chia hết cho 4 và không chia hết cho 100, hoặc

Chia hết cho 400.



```python
n = int(input("nhập năm: "))
if (n % 4 ==0 and n % 100 != 0) or (n % 400 == 0):
    print("đây là năm nhuận")
else:
    print("đây không phải năm nhuận")
```

6. Tính tổng các số từ 1 đến n
Nhập số nguyên dương n, tính tổng từ 1 đến n.

Gợi ý: dùng vòng lặp for hoặc công thức.


```python
n = int(input("nhập một số nguyên dương: "))
tong = 0
for i in range(1,n+1):
    tong += i
print("tổng của các số từ 1 đến n là: ",tong)
```

7. Đếm số nguyên âm, dương, bằng 0
Nhập vào danh sách 10 số nguyên, đếm xem có bao nhiêu số âm, số dương và số bằng 0.




```python
danh_sach = []
for i in range(10):
   n = int(input(f"nhập số thứ {i+1}: "))
   danh_sach.append(n)
am = 0
duong = 0
khong = 0
for n in danh_sach:
   if n > 0:
     duong += 1
   elif n== 0:
     khong += 1
   else:
     am += 1
print("số dương: ", duong)
print("số âm: ", am)
print("số bằng 0: ", khong)
```

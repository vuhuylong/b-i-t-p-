```python
x = float(input("nhập một số nguyên "))
if x > 0:
     print("x là một số nguyên dương")
     y = x**0.5
     print(f"căn bậc 2 của x là {y: .3f}")
```

```python
s = input("enter 3 real number(separated by space):")
s = s.split()
a = float(s[0])
b = float(s[1])
c = float(s[2])
print(f"a = {a}, b = {b}, c = {c}")
# a**2*x+b*x+c
```

```python
n = int(input("Nhập 1 số nguyên dương: "))
s = 0
for i in range(0,n + 1):
    s += i
print(f"tổng của số 0 đến {n} là {s}")
```

```python
n = int(input("nhập 1 số nguyên dương:"))
print(n)
s = 0
i = 2
while i < n:
      s += i
      i += 2
print(f"tổng các số nguyên dương nhỏ hơn {n} là {s}")
```
